```python
import requests
```

```python
def get_most_similar_aliases(phrase, topn = 20):
    url_to_hit = "https://172.29.93.122/most_similar_single"
    params = {'query_word': phrase, 'topn': topn}
    resp_json = requests.get(url_to_hit, params, verify=False).json()
    return resp_json[0]['value']
```

```python
def get_fixed_len_list_sim(phrase, max_len_list = 25):
    ret_list = get_most_similar_aliases(phrase, topn = max_len_list)
    if(type(ret_list)==list):
        trim_list = ret_list[:max_len_list]
    else:
        trim_list = []

    fin_list = trim_list+(max_len_list - len(trim_list))*[['NA',0.0]]
    return fin_list
```
